```python
#Swap two numbers using third variable
num1=int(input("Enter first number: ") )
num2=int(input("Enter second number: ") )
print(f"Before swapping: num1={num1} and num2={num2}")
temp=num1
num1=num2
num2=temp
print(f"After swapping: num1={num1} and num2={num2}")
```

```python
#Check weather a number is power of 2 or not
num=int(input("Enter a number: ") )
if num>0 and (num & (num-1))==0:
    print(f"{num} is a power of 2.")
else:
    print(f"{num} is not a power of 2.")
```

```python
#Find the greatest of three numbers
num1=int(input("Enter first number: ") )
num2=int(input("Enter second number: ") )
num3=int(input("Enter third number: ") )
if num1>=num2 and num1>=num3:
    greatest=num1
elif num2>=num1 and num2>=num3:
    greatest=num2
else:
    greatest=num3
print(f"The greatest of {num1}, {num2} and {num3} is: {greatest}")
```
