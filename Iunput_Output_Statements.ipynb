```python
# Print the size of char,int,float and double
character=input("Enter the character: ")
integer=int(input("Enter the integer: "))
float_number=float(input("Enter the float: "))
print(f"Value :{character} Size:{len(character)}\nValue :{integer} Size: {len(str(integer))}\nValue :{float_number} Size: {len(str(float_number))} ")
```
